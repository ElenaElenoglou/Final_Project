**Content_based exploration**

```python
# Proportion de sessions par genre de contenu
count_genre = Salto_V1.groupby(['tag_genre'])['pid'].count().reset_index()
count_genre= count_genre.sort_values('pid', ascending=False)
count_genre = count_genre.rename(columns = {'user_count':'user_count'})
count_genre.loc[:, 'proportion'] = 100*count_genre['pid']/count_genre['pid'].sum()
count_genre
```

```python
# Let's see some visualisations
data = count_genre
fig = px.bar(data, x=data.tag_genre, y=data.proportion, title="Proportion de sessions par genre de contenu", color=data.index)
fig.show()
```

```python
# Proportion de sessions par sous genre de contenu
count_ss_genre = Salto_V1.groupby(['tag_sous_genre'])['pid'].count().reset_index()
count_ss_genre= count_ss_genre.sort_values('pid', ascending=False)
count_ss_genre= count_ss_genre.rename(columns = {'user_count':'user_count'})
count_ss_genre.loc[:, 'proportion'] = 100*count_ss_genre['pid']/count_ss_genre['pid'].sum()
count_ss_genre
```

```python
# Let's see some visualisations
data = count_ss_genre.head(10)
fig = px.bar(data, x=data.tag_sous_genre, y=data.proportion, title="Proportion de sessions par sous genre de contenu", color=data.index)
fig.show()
```

```python
# Proportion de sessions par format de vidéo
video_type = Salto_V1.groupby(['video_format'])['pid'].count().reset_index()
video_type= video_type.sort_values('pid', ascending=False)
video_type = video_type.rename(columns = {'user_count':'user_count'})
video_type.loc[:, 'proportion'] = 100*video_type['pid']/video_type['pid'].sum()
video_type
```

```python
# Let's see some visualisations
data = video_type
fig = px.bar(data, x=data.video_format, y=data.proportion, title="Proportion de sessions par type de format de video", color=data.index)
fig.show()
```

```python
# Proportion de sessions par format et par genre
video_type_cat = Salto_V1.groupby(['video_format','tag_genre'])['pid'].count().reset_index()
video_type_cat= video_type_cat.sort_values('pid', ascending=False)
video_type_cat= video_type_cat.rename(columns = {'user_count':'user_count'})
video_type_cat.loc[:, 'proportion'] = 100*(video_type_cat['pid']/(video_type_cat['pid'].sum()))
video_type_cat
```

```python
# Let's see some visualisations
data = video_type_cat
fig = px.bar(data, x=data.video_format, y=data.proportion, color=data.tag_genre, title="Proportion de sessions par type de format et par genre")
fig.show()
```

```python
# Proportion de sessions par format et par sous genre
video_type_ss_genre = Salto_V1.groupby(['video_format','tag_sous_genre'])['pid'].count().reset_index()
video_type_ss_genre= video_type_ss_genre.sort_values('pid', ascending=False)
video_type_ss_genre= video_type_ss_genre.rename(columns = {'user_count':'user_count'})
video_type_ss_genre.loc[:, 'proportion'] = 100*(video_type_ss_genre['pid']/(video_type_ss_genre['pid'].sum()))
video_type_ss_genre
```

```python
# Let's see some visualisations
data = video_type_ss_genre.head(10)
fig = px.bar(data, x=data.video_format, y=data.proportion, color=data.tag_sous_genre, title="Proportion de sessions par type de format et par sous genre")
fig.show()
```

```python
# Age moyen par genre
age_genre = Salto_V1.groupby(['tag_genre'])['age'].mean().reset_index()
age_genre.loc[:, ['age']]=round(age_genre.loc[:, ['age']], 0).astype(int)
age_genre=age_genre.sort_values('age', ascending=False)
age_genre= age_genre.rename(columns = {'age':'age_mean'})
#age_genre.loc[:, 'proportion'] = 100*(age_genre['count']/(age_genre['count'].sum()))
age_genre
```

```python
# Let's see some visualisations
data = age_genre
fig = px.bar(data, x=data.tag_genre, y=data.age_mean, color=data.index, title="Age moyen par genre")
#fig.update_layout(barmode='stack')
fig.show()
```

```python
# Distrbution age par genre = Divertissement
age_genre = Salto_V1.groupby(['age', 'tag_genre'])['pid'].count().reset_index()
age_genre= age_genre[age_genre['tag_genre']=='divertissement']
#age_genre.loc[:, ['age']]=round(age_genre.loc[:, ['age']], 0).astype(int)
age_genre=age_genre.sort_values('age', ascending=True)
#age_genre= age_genre.rename(columns = {'age':'age'})
#age_genre.loc[:, 'proportion'] = 100*(age_genre['count']/(age_genre['count'].sum()))
age_genre
```

```python
# Let's see some visualisations
data = age_genre.loc[age_genre['age'] < 85, :]
fig = px.bar(data, x=data.age, y=data.pid, color=data.index, title="Distribution de l'age pour le divertissement")
#fig.update_layout(barmode='stack')
fig.show()
```

```python
# Durée moyenne de vidéo par genre (en minutes)
video_convert_genre= Salto_V1.groupby(['tag_genre'])['video_duration'].mean().reset_index()
#age_genre=age_genre.sort_values('pid', ascending=False)
video_convert_genre= video_convert_genre.rename(columns = {'video_duration':'mean'})
video_convert_genre['video_duration_mean_in_min'] = video_convert_genre['mean']/60
video_convert_genre
```
